# Analyse du churn avec un modèle de régression logisitique

```r
dataset <- read.csv("customer_churn.csv")
head(dataset)
```

```r
summary(dataset)
```

```r
reg_logistique = glm(Churn ~ Age + Total_Purchase + Num_Sites + Years + Account_Manager, family = binomial, data = dataset)
summary(reg_logistique)
```

## Réaliser des prédictions avec le modèle de régression logisitique

```r
# Sélectionner les variables utiles
dataset_pred <- dataset %>% select(Age, Num_Sites, Years, Total_Purchase, Account_Manager)
head(dataset_pred)
```

```r
# Prédire avec le modèle de régression logisitique
predictions <- predict(reg_logistique, dataset_pred, type = "response")
predictions
```

```r
# Arrondir à 0 ou 1 en fonction d'un seuil de 0.5
predictions_class <- ifelse(predictions >= 0.5, 1, 0)
predictions_class
```

```r
# Moeynne des bonnes prédictions
mean(dataset$Churn == predictions_class)
```
